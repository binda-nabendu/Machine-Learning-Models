```python
import math
pic, cordinate = plt.subplots(nrows = 3, ncols = 2)

x1 = {'v1': x, 'v2': [xi ** 2 for xi in x], 'v3':[math.sqrt(xi) for xi in x]}
y1 = {'w1': y, 'w2': [yi ** 3 for yi in y], 'w3': y}

# Iterate over the subplots and data dictionaries
for (subplot, (x_key, x_data), (y_key, y_data)) in zip(cordinate.flat, x1.items(), y1.items()):
    subplot.plot(x_data, y_data)
    subplot.set_title(f'{x_key} vs {y_key}')
```
