```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.distplot(tips['total_bill']) #Relative frequency distribution: provide percentage
```

```python
sns.displot(tips['total_bill'], kde = True) #Regular frequency distribution: provide count
```

```python
sns.distplot(tips['total_bill'], kde = False)
```

```python
sns.distplot(tips['total_bill'], kde = False, bins= 30)
```

```python
sns.jointplot(x='total_bill', y = 'tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y = 'tip', data=tips, kind='hex', gridsize = 40)
```

```python
sns.jointplot(x='total_bill', y = 'tip', data=tips, kind='reg') #kind= 'reg' means regression
```

```python
import matplotlib.pyplot as plt
```

```python
sns.jointplot(x='total_bill', y = 'tip', data=tips, kind='kde', palette=['red', 'green', 'blue'])
plt.show()
```

```python
sns.jointplot(x='total_bill', y = 'tip', data=tips, kind='kde', palette=['red', 'green', 'blue'], fill = 'True')
plt.show()
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'day')# hue parameter categorized depend on provided column name
```

```python
sns.pairplot(tips)
sns.pairplot(tips, hue='sex', palette = ['red', 'green', 'blue'])
plt.show()
```

```python
sns.pairplot(tips)
sns.pairplot(tips,vars = ['tip', 'size'], hue='sex', palette = ['red', 'green', 'blue'], diag_kind = 'hist')
plt.show()
```

```python
sns.pairplot(tips)
sns.pairplot(tips,vars= ['total_bill', 'size'], hue='sex', palette = ['red', 'green'])
plt.show()
```

```python
sns.rugplot(data = tips['total_bill'])
```

```python
sns.rugplot(data = tips, x = 'size', y = 'total_bill', hue = 'sex')
```
