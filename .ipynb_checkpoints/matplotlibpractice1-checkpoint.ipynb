```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

#Functional

```python
plt.plot(x, y)
# plt.show()
```

```python
plt.plot(x, y, 'r-')
```

```python
plt.plot(x, y)
plt.xlabel("x label")
plt.ylabel("Y label")
plt.title("Demo")
```

```python
plt.subplot(1, 3, 1)
plt.plot(x, y, 'g')

plt.subplot(1, 3, 2) # argument : number of row , number of column, plot number
plt.plot(y, x, 'y')

plt.subplot(1, 3, 3)
plt.plot(x, y + 2, 'w')
```

```python
#Object oriented
oop_fig = plt.figure()
axis = oop_fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis.plot(x, y)
axis.set_xlabel('X Label')
axis.set_ylabel('X Label')
axis.set_title('X Label')
```

```python
oop_fig2 = plt.figure()
axis1 = oop_fig2.add_axes([0.1, 0.1, 0.8, 0.8])
axis2 = oop_fig2.add_axes([0.2, 0.3, 0.4, 0.3])# 20% from left, 30% from bottom, 40% width from canvas, 30 height from canvas
axis3 = oop_fig2.add_axes([0.3, 0.4, 0.4, 0.3])# 20% from left, 30% from bottom, 40% width from canvas, 30 height from canvas
axis1.plot(x, y)
axis3.plot(y, x)
axis2.plot(x, y)

axis1.set_title('X Y plot')
axis3.set_title('y X plot')
```

```python
oop_fig2,axes = plt.subplots(nrows=1, ncols=2)
# axes.plot(x, y)
# plt.tight_layout()
# for current in axes:
#     current.plot(x, y)
# axes
axes[0].plot(x, y)
```
