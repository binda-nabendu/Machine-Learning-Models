```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.distplot(tips['total_bill']) #Relative frequency distribution: provide percentage
```

```python
sns.displot(tips['total_bill'], kde = True) #Regular frequency distribution: provide count
```

```python
sns.distplot(tips['total_bill'], kde = False)
```

```python
sns.distplot(tips['total_bill'], kde = False, bins= 30)
```

```python
sns.jointplot(x='total_bill', y = 'tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y = 'tip', data=tips, kind='hex', gridsize = 40)
```

```python
sns.jointplot(x='total_bill', y = 'tip', data=tips, kind='reg') #kind= 'reg' means regression
```

```python
sns.jointplot(x='total_bill', y = 'tip', data=tips, kind='kde')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex')# hue parameter categorized depend on provided column name
```

```python
sns.pairplot(tips)
sns.pairplot(tips, hue='sex', palette = ['red', 'green', 'blue'])
```

```python
sns.rugplot(tips['total_bill'])
```

Catagorical Data plotting

```python
sns.barplot(x= 'sex', y = 'total_bill', data = tips)
```

```python
import numpy as np
```

```python
sns.barplot(x= 'sex', y = 'total_bill', data = tips, estimator= np.std)
```

```python
sns.countplot(x='sex', data = tips)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
sns.violinplot(x='day', y = 'total_bill', data = tips)
```

```python
sns.violinplot(x='day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.violinplot(x='day', y = 'total_bill', data = tips, hue = 'sex', split= True)
```

```python
sns.stripplot(x='day', y = 'total_bill', data= tips)
```

```python
sns.stripplot(x='day', y = 'total_bill', data= tips, jitter = False)
```

```python
sns.stripplot(x='day', y = 'total_bill', data= tips, jitter = True, hue = 'sex')
```

```python
sns.stripplot(x='day', y = 'total_bill', data= tips, jitter = True, hue = 'sex', dodge = True) # 'split' work like 'dodge'
```

```python
sns.swarmplot(x='day', y='total_bill', data = tips)
```

```python
sns.violinplot(x='day', y='total_bill', data = tips)
sns.swarmplot(x='day', y='total_bill', data = tips, color='yellow')
```
