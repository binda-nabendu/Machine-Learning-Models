```python
import numpy as np
x = np.linspace(0, 100, 30)
y = x ** 2
```

```python
plt.plot(x, y)
plt.title("This is demo graph")
plt.xlabel('Here present x')
plt.ylabel('Here present y')
plt.show()
```

```python
plt.subplot(1, 3, 1)
plt.plot(x, y, color = 'red')

plt.subplot(1, 3, 3)
plt.plot(y, x, color = 'green')
```

```python
oop_figure = plt.figure()
graph_paper = oop_figure.add_axes([.25, .15, 0.2, 0.6])
graph_paper.plot(x,y)
graph_paper.set_xlabel('x')
graph_paper.set_ylabel('y')
graph_paper.set_title("OO Plot")
```

```python
oop_figure2 = plt.figure()
main_plot = oop_figure2.add_axes([.25, .15, .9, .7])

subplot1 = oop_figure2.add_axes([.3, .2, .4, .3])

subplot2 = oop_figure2.add_axes([.73, .54, .4, .3])

main_plot.plot(x, y, 'r')

subplot1.plot(x, x ** 3, 'r')

subplot2.plot(y, x, 'g')
```

```python
import math
```

```python
pic, cordinate = plt.subplots(nrows = 4, ncols = 3)

all_x_array = {'first all x': x, 'second all x': [element ** 2 for element in x], 'third element x':y}
all_y_array = {'first all y': y, 'second all y': [yi - 2 for yi in y], 'third element of y':[math.log(xi ** 4) for xi in x]}

for(target_plot, (x_key, x_value), (y_key, y_value)) in zip(cordinate.flat, all_x_array.items(), all_y_array.items()):
    target_plot.plot(x_value, y_value)
    target_plot.set_xlabel(x_key)
    target_plot.set_ylabel(y_key)
    target_plot.set_title(f"{x_key} and {y_key}")
plt.tight_layout()
```
