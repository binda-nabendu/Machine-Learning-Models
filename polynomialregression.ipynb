```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_function = PolynomialFeatures(degree= 2, include_bias= False)
x_in_polynomial = polynomial_function.fit_transform(x)
```

```python
y.dtype
```

```python
reg = LinearRegression()
# reg.fit(x, y)
reg.fit(x_in_polynomial, y)
```

```python
# x_vals = np.linspace(0, 1, 100).reshape(-1, 1)
```

```python
x_vals = np.linspace(-2, 2, 100).reshape(-1, 1)
```

```python
poly_x_val_test = polynomial_function.transform(x_vals)
```

```python
y_vals = reg.predict(poly_x_val_test)
```

```python
plt.scatter(x, y)
plt.plot(x_vals, y_vals, color = 'yellow')
plt.show
```
