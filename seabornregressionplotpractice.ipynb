```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x = 'total_bill', y='tip', data = tips, hue = 'sex')
```

```python
tips.head()
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s':200}) # s:200 means it increase the size of all point
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row = 'time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row = 'time', hue = 'sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue = 'sex', aspect = 0.7, height = 8)
```
