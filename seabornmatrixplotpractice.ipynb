```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
import pandas as pd
```

```python
tips.head()
```

```python
flights.head()
```

```python
tipsnum = tips.select_dtypes(include='number')
```

```python
tipsnum
```

```python
tip_rel = tipsnum.corr()
```

```python
sns.heatmap(tip_rel)
```

```python
sns.heatmap(tip_rel, annot=True)
```

```python
sns.heatmap(tip_rel, annot=True, cmap='coolwarm')
```

```python
flights
```

```python
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(flights_pivot)
```

```python
sns.heatmap(flights_pivot, cmap = 'magma')
```

```python
sns.heatmap(flights_pivot, cmap = 'magma', linecolor= 'yellow', linewidths= .5)
```

```python
sns.clustermap(flights_pivot)  # to get that install scipy
```

```python
sns.clustermap(flights_pivot, cmap = 'coolwarm', standard_scale = 1)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(plt.kdeplot)
```

```python
g = sns.FacetGrid(data= tips, col = 'time', row = 'smoker')
```

```python
g = sns.FacetGrid(data= tips, col = 'time', row = 'smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data= tips, col = 'time', row = 'smoker')
g.map(sns.scatterplot, 'total_bill', 'tip')
```
