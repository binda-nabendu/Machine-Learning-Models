```python
pic, cordinate = plt.subplots(nrows = 3, ncols = 2)

x1 = {'first plot x': x, 'second plot x': [xi ** 2 for xi in x], 'third plot x': [xi ** 2 for xi in x]}
y1 = {'first plot y': y, 'second plot y': [yi - 2 for yi in y], 'third plot y': [math.log(yi, 10) for yi in y]}

for(subplot,(x_key, x_value), (y_key, y_value)) in zip(cordinate.flat, x1.items(), y1.items()):
    subplot.plot(x_value, y_value)
    subplot.set_title(x_key + " and " + y_key)
    subplot.set_xlabel(f"{x_key}")
    subplot.set_ylabel(y_key)
plt.tight_layout()
```

all property of scatter plot
plt.scatter(
    x,             # x-coordinates of data points
    y,             # y-coordinates of data points
    s=marker_size, # Size of markers (default is 20)
    c='color',     # Color of markers (can be a single color or an array of colors)
    marker='marker_style',  # Marker style (default is 'o' for circles)
    alpha=opacity,  # Opacity of markers (0.0 transparent through 1.0 opaque)
    label='legend_label',  # Label for legend entry
    cmap='colormap',  # Colormap for mapping numeric values to colors
    norm='normalization',  # Normalization for the colormap
    edgecolors='edge_color',  # Color of marker edges
    linewidths='line_width',   # Width of marker edges
    **kwargs        # Additional keyword arguments for customization
)
Here's a brief explanation of these parameters:

x and y: The data points' x and y coordinates.
s: Size of markers (default is 20). Can be a single value or an array for varying sizes.
c: Color of markers. You can provide a single color (like 'red') or an array of colors.
marker: Marker style (default is 'o' for circles). You can use various marker styles like 's' (square), '^' (triangle), etc.
alpha: Opacity of markers, ranging from 0.0 (transparent) to 1.0 (opaque).
label: Label for the legend entry associated with this scatter plot.
cmap and norm: Colormap and normalization for mapping numeric values to colors.
edgecolors: Color of marker edges.
linewidths: Width of marker edges.

```python
# Generate some sample data
data = np.random.randn(1000)  # Generating random data

# Create a histogram
plt.hist(data, bins=20, edgecolor='red')  # 'bins' parameter controls the number of bins
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Generate some sample data
data1 = np.random.normal(0, 1, 500)
data2 = np.random.normal(2, 1, 500)

# Create histograms for both datasets
plt.hist(data1, bins=20, alpha=0.5, label='Data 1')
plt.hist(data2, bins=20, alpha=0.5, label='Data 2')

# Adding decorations
plt.title('Histogram Comparison')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()  # Adding legend
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 10, 20)
y = np.sin(x)

plt.plot(x, y, marker='o', label='Data Points')

for i, (xi, yi) in enumerate(zip(x, y)):
    plt.text(xi, yi, f'({xi:.2f}, {yi:.2f})', ha='center', va='bottom') #horizental ha, vertical va

plt.title('Line Plot with Data Labels')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
